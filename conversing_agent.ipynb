```python

# import os, getpass
from openai import OpenAI

# from google.colab import userdata
# base_url = userdata.get('LITELLM_BASE_URL')
# api_key = userdata.get('LITELLM_API_KEY')

base_url="your base_url"
api_key="your api_key"
MODEL = "openai/gpt-4o-mini"   # the model alias your LiteLLM proxy serves

client = OpenAI(
    base_url=base_url,
    api_key=api_key,
    timeout=600.0, # Increase timeout to 10 minutes
)
print("LiteLLM client ready | base_url:", base_url, "| model:", MODEL)
```

```python
client.chat.completions.create(
          model=MODEL,
          temperature=0.7,
          messages=[{"role":"system", "content":"You are a question answering chatbot, greet user well"},
                    { "role": "user","content": "hi i am rasmi, who are you?" }]
    )
```

```python
message_history = []

def ask(prompt, message_history,
        temperature=0.7,
        model=MODEL):

    system = "You are a question answering chatbot, greet user well"

    print("Calling API...")
    message_history.append({
        "role": "user",
        "content": prompt
    })

    r = client.chat.completions.create(
          model=model,
          temperature=temperature,
            messages=[
            {"role": "system", "content": system}
        ]+message_history
    )



    answer = r.choices[0].message.content



    message_history.append({
        "role": "assistant",
        "content": answer
    })

    return answer, r.usage
```

```python

temperature=float(input("Enter temperature: "))
# temp btw 0 to 2
```

```python
while True:
  userInput=input("You: ")
  if userInput.lower() == "exit":
        break
  print("request was send")
  answer, usage = ask(
        userInput,
        message_history,
        temperature=temperature)
  print("Agent:", answer)

  print("Prompt Tokens:", usage.prompt_tokens)
  print("Completion Tokens:", usage.completion_tokens)
  print("Total Tokens:", usage.total_tokens)
```
